```python
import urllib.request
from datetime import date
from bs4 import BeautifulSoup
from unidecode import unidecode
from os.path import isfile
import pandas as pd
```

```python
DATA_PATH = '../data/'
BILLBOARD_RAW_PATH = 'billboard_raw/'
BILLBOARD_RAW_PATH2 = 'billboard_raw2/'
```

```python
def fetchList(year):
    # fetch initial raw from billbaord chart 
    bbSite_before = 'https://www.billboard.com/charts/year-end/'
    bbSite_after = '/hot-100-songs'
    response = urllib.request.urlopen(bbSite_before + str(year) + bbSite_after)
    print(bbSite_before + str(year) + bbSite_after)
    html = response.read()
    soup = BeautifulSoup(html, 'html.parser')

    # titles 
    title_list = []
    titles = soup.findAll("div", {"class": "ye-chart-item__title"})
    for title in titles:
        title_list.append(unidecode(title.text).strip('\n'))

    # artists
    artist_list = []
    artists = soup.findAll("div", {"class": "ye-chart-item__artist"})
    for artist in artists:
        artist_list.append(unidecode(artist.text).strip('\n'))
    return title_list,artist_list
```

```python
def add_attributes(df,year):
    position_list = df['position']
    index_list = list(range(0,len(position_list)))
    year_list = [year]*len(position_list)
    title_list = df['title']
    artist_list = df['artist']
    dec_list = [dec_calculate(year)]*len(position_list)
    song_key_list = genrate_song_key(year_list,position_list)
    d = {'index': index_list, 'position':position_list,'year': year_list, 'artist': artist_list, 'title': title_list, 'decade': dec_list, 'song_key': song_key_list}
    df = pd.DataFrame(data=d) 
    df = df[['index','position','year','title','artist','decade','song_key']]
    return df
    
    
```

```python
def dec_calculate(year):
    return divmod(year,10)[0]*10
```

```python
def genrate_song_key(year_list,position_list):
    song_key_list = []
    for i in range(len(year_list)):
        song_key_list.append(str(year_list[i])+"_"+str(position_list[i]))
    return song_key_list
```

```python
# NOTE: Billboard official website only provides from year 2006
start_year = 2006
curr_year = date.today().year
# iterate to the current year - 1
for y in range(start_year, curr_year):
    fname = DATA_PATH + BILLBOARD_RAW_PATH + str(y) + 'hot100.csv'
#     if isfile(fname):
#         continue
#     else:
#         fp = open(fname, 'w')
#         fp.close()

    title_list = []
    artist_list = []
    title_list, artist_list = fetchList(y)
    
    rank_list = list(range(1,len(title_list)+1))

    d = {'position':rank_list, 'artist': artist_list, 'title': title_list}
    df = pd.DataFrame(data=d) 
    df = df[['position', 'artist', 'title']]
    # add helper attributes
    df = add_attributes(df,y)
    df.to_csv(fname, sep='@')
    display(df)
```

```python
bbSite = 'http://billboardtop100of.com/'

start_year = 1970
curr_year = 2006

for y in range(start_year, curr_year):
    fname = DATA_PATH + BILLBOARD_RAW_PATH2 + str(y) + 'hot100.csv'
#     if isfile(fname):
#        continue
#     fp = open(fname, 'w')
#     fp.close()
    rank = 1
    title_list = []
    artist_list = []
    rank_list = []
    response = urllib.request.urlopen(bbSite + str(y) + '-2/')
    html = response.read()
    soup = BeautifulSoup(html, 'html.parser')

    if soup.table:
        rows = soup.table.findChildren(['tr'])
        for row in rows:
            cells = row.findChildren('td')
            out_line = []
            for cell in cells:
                out_line.append(unidecode(cell.text))
            out_str = '@'.join(out_line)
            out_list = out_str.split('@')
            rank_list.append(rank)
            rank += 1
            artist_list.append(out_list[1])
            title_list.append(out_list[2])
        d = {'position':rank_list, 'artist': artist_list, 'title': title_list}
        df = pd.DataFrame(data=d) 
        df = df[['position', 'artist', 'title']]
        # add helper attributes
        df = add_attributes(df,y)
        df.to_csv(fname, sep='@')
        display(df)
```
