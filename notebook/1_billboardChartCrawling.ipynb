```python
import urllib.request
from datetime import date
from bs4 import BeautifulSoup
from unidecode import unidecode
from os.path import isfile
import pandas as pd
```

```python
DATA_PATH = '../data/'
BILLBOARD_RAW_PATH = 'billboard_raw/'
```

```python
def fetchList(year):
    # fetch initial raw from billbaord chart 
    bbSite_before = 'https://www.billboard.com/charts/year-end/'
    bbSite_after = '/hot-100-songs'
    response = urllib.request.urlopen(bbSite_before + str(year) + bbSite_after)
    print(bbSite_before + str(year) + bbSite_after)
    html = response.read()
    soup = BeautifulSoup(html, 'html.parser')

    # titles 
    title_list = []
    titles = soup.findAll("div", {"class": "ye-chart-item__title"})
    for title in titles:
        title_list.append(unidecode(title.text).strip('\n'))

    # artists
    artist_list = []
    artists = soup.findAll("div", {"class": "ye-chart-item__artist"})
    for artist in artists:
        artist_list.append(unidecode(artist.text).strip('\n'))
    return title_list,artist_list
```

```python
# NOTE: Billboard official website only provides from year 2006
start_year = 2006
curr_year = date.today().year
# iterate to the current year - 1
for y in range(start_year, curr_year):
    fname = DATA_PATH + BILLBOARD_RAW_PATH + str(y) + 'hot100.csv'
    if isfile(fname):
        continue
    else:
        fp = open(fname, 'w')
        fp.close()

    title_list = []
    artist_list = []
    title_list, artist_list = fetchList(y)
    
    rank_list = list(range(1,len(title_list)+1))

    d = {'rank':rank_list, 'artist': artist_list, 'title': title_list}
    df = pd.DataFrame(data=d) 
    df = df[['rank', 'artist', 'title']]
    df.to_csv(fname, sep='@')
    display(df)
```
