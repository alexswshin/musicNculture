```python
import requests
from bs4 import BeautifulSoup
import json
import types
from os import listdir
from os.path import isfile, join
from unidecode import unidecode
import pandas as pd
import codecs
from datetime import date
```

```python
DATA_PATH = '../data/'
BILLBOARD_RAW_PATH = 'billboard_raw/'
BILLBOARD_RAW_PATH2 = 'billboard_raw2'
BILLBOARD_LYRIC_PATH = 'billboard_lyrics/'
BILLBOARD_LYRIC_PATH2  = 'billboard_lyrics2/'
```

```python
df_total = pd.DataFrame() #creates a new dataframe that's empty 
       
start_year = 1970
curr_year = 2006
    
for y in range(start_year, curr_year):
    file = str(y) + 'hot100.csv'
    fname = DATA_PATH+BILLBOARD_LYRIC_PATH2+file
    print(fname)
    df = pd.read_csv(fname,sep='@')
    df_total = df_total.append(df, ignore_index = True)

    
    
start_year = 2006
curr_year = date.today().year
    
for y in range(start_year, curr_year):
    file = str(y) + 'hot100.csv'
    fname = DATA_PATH+BILLBOARD_LYRIC_PATH+file
    print(fname)
    df = pd.read_csv(fname,sep='@')
    df_total = df_total.append(df, ignore_index = True)

df_total = df_total[['position','year','decade','title','artist','lyric']]
df_total.to_csv(DATA_PATH+"billboard_1970_current.csv", sep='@') 
display(df_total)
 
```
